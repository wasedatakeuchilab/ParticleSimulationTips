# 補間

数値計算における補間とは既知の離散点から連続値関数を構成して未知の離散点の値を算出する方法である。

補間には様々な手法があるが、このノートブックでは基本的な線形補間とBSpline曲線による補間をまとめる。

```julia
using Interpolations
using Plots

# 粗いサインカーブを作成 -----------------------------------------
x = range(0, 2π, length = 10)
y = sin.(x)
plt = plot(x, y, marker = :circle)
```

```julia
# 線形補間を実行 ------------------------------------------------
itp = linear_interpolation(x, y)
x_fine = range(0, 2π, length = 100)
scatter(plt, x_fine, itp.(x_fine))
```
